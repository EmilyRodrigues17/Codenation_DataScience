```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

## Explorando os dados

```python
df_train = pd.read_csv('train.csv')
```

```python
df_test = pd.read_csv('test.csv')
```

```python
df_test.head(2)
```

```python
df_train.head(2)
```

## Deixando so as colunas necessárias em df_train

```python
colunas_teste = list(df_test.columns)
```

```python
colunas_teste.append('IN_TREINEIRO')
```

```python
df_train = df_train[colunas_teste]
```

```python
df_train.head(2)
```

```python
df_train.describe()
```

```python
df_train_corr = df_train.corr()
df_train_corr
```

```python
corr_treino = df_train_corr.loc['IN_TREINEIRO']
corr_treino
```

```python
variaveis = corr_treino[(corr_treino >= 0.25) | (corr_treino <= -0.25)]
```

```python
variaveis = variaveis.index
```

```python
variaveis
```

## --- Criação do modelo ---

```python
df_train = df_train[variaveis]
```

```python
df_train.head(2)
```

```python
y_train = df_train['IN_TREINEIRO']
X_train = df_train.drop(columns = 'IN_TREINEIRO')
```

```python
X_test = df_test[X_train.columns]
```

```python
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
classifier = RandomForestClassifier(n_estimators = 100, random_state = 42)
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred_test = classifier.predict(X_test)
```

```python
df_answer = pd.DataFrame()
df_answer['NU_INSCRICAO'] = df_test['NU_INSCRICAO']
df_answer['IN_TREINEIRO'] = y_pred_test
df_answer
```

```python
df_answer.to_csv('answer.csv', index=False, header=True)
```
